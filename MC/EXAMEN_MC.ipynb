**PARTE1 | CARGA Y LIMPIEZA**

```python
#librerioas
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
url='https://raw.githubusercontent.com/mwaskom/seaborn-data/master/mpg.csv'
df= pd.read_csv('https://raw.githubusercontent.com/mwaskom/seaborn-data/master/mpg.csv')
```

```python
print(df.head(6))
print("datos")
```

```python
mediana_hp = np.nanmedian(df['horsepower'])
df['horsepower'] = df['horsepower'].fillna(mediana_hp)
df['horsepower'].isnull().sum()
df['relacionPP'] = df['weight'] / df['horsepower']
df.head()
```

**PARTE 2 | GRAIFCOS**

Histograma: Muestra la distribución de la aceleración (acceleration) de los vehículos.

```python
plt.hist(df['acceleration'])
plt.xlabel("Aceleración")
plt.ylabel("")
plt.title("Distribución de la Aceleración")
plt.grid(axis='y')
plt.show()
```

Gráfico de Dispersión (Scatter): Grafica el Desplazamiento del motor (displacement) en el eje X versus los Caballos de fuerza (horsepower) en el eje Y.

```python
plt.scatter(df['displacement'], df['horsepower'])
plt.title('Displacement vs. Horsepower')
plt.xlabel('desplazamiento del motor')
plt.ylabel('Horsepower ')
plt.grid(True)
plt.show()
```

Gráfico de Barras: Muestra el promedio de rendimiento (mpg - millas por galón) agrupado por el país de origen (origin). Pista: Tendrás que agrupar los datos primero con Pandas.

```python

Agrupacion = df.groupby('origin')['mpg'].mean()
plt.bar(Agrupacion.index, Agrupacion.values, color='purple', edgecolor="purple")
plt.title('Millas por galon-origen')
plt.xlabel('Origen')
plt.ylabel('MPG')
plt.xticks(rotation=0)
plt.grid(axis='y')
plt.show()
```
